## Example destination notebook
This notebook is an example workflow that shows how the [scrapbook](https://nteract-scrapbook.readthedocs.io/) library from [nteract](https://nteract.io) could be used in a workflow. This notebook reads the results of a set of previously executed notebooks.

````python
source_dir = "tickers"
````

````python
import pandas as pd

import matplotlib.pyplot as plt

import scrapbook as sb
````

In the [previous notebook](https://github.com/wrighter/python_blogposts/tree/main/tools/scrapbook_example_source.ipynb) data was generated, a graph was plotted, and several values persisted into a notebook using scrapbook. We also ran a number of notebooks using papermill. Now, the notebook are retrieved by loading them all using scrapbook and accessing the values in this notebook. First, we need to read the notebooks.

````python
sbook = sb.read_notebooks(source_dir)
````

We can now display all the data previously glued in the notebook by interrogating the scraps. For this example, we will check the ```close_to_ath``` value for truth, and display details if the price is near the ATH.

````python
for nb in sbook.notebooks:
    print(f"{nb.scraps['symbol'].data: <5} {nb.scraps['distance'].data * 100: .2f}%")
    if nb.scraps['close_to_ath'].data:
        display(nb.scraps['chart'].display['data'], raw=True)        
````

Another way to view display data is to ```reglue``` a scrap into the current notebook.

````python
nb.reglue("chart")                         # image, displayed
````
